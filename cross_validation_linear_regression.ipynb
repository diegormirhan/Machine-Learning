```python
from sklearn.linear_model import LinearRegression

import pandas as pd
import numpy as np
```

```python
# Carregando o dataset de diabetes e exibindo as 5 primeiras linhas
diabetes_df = pd.read_csv("diabetes.csv")
diabetes_df.head()
```

```python
# Separando features (X) e variável alvo (y)
# X = todas as colunas EXCETO glucose
# y = coluna glucose (o que queremos prever)
X = diabetes_df.drop("glucose", axis=1).values
y = diabetes_df["glucose"].values
```

```python
from sklearn.model_selection import cross_val_score, KFold

# Dividindo os dados em 6 partes (folds) para cross-validation
kf = KFold(n_splits=6, shuffle=True, random_state=42)
model =  LinearRegression()

# Treina e avalia o modelo 6 vezes, cada vez usando uma parte diferente como teste
cv_results = cross_val_score(model, X, y, cv=kf)
```

```python
# Score R² de cada um dos 6 folds
print(cv_results)

# Média e desvio padrão dos scores (resumo geral do desempenho)
print(np.mean(cv_results), np.std(cv_results))

# Intervalo de confiança de 95%: mostra a faixa real (min e max) onde o score do modelo se encontra
print(np.quantile(cv_results, [0.025, 0.975]))
```
