```python
import pandas as pd
import numpy as np

```

```python
diabetes_df = pd.read_csv("diabetes.csv")
diabetes_df.head()
```

```python
X = diabetes_df.drop("glucose", axis=1).values
y = diabetes_df["glucose"].values

# Previsões com base em uma única variável independente
X_bmi = X[:, 3] # todas as linhas mas somente a 4ª coluna
print(y.shape, X_bmi.shape) # (768,) todas as linhas mas somente uma coluna

X_bmi = X_bmi.reshape(-1, 1) # quantas linhas forem necessárias mas com 1 coluna
print(X_bmi.shape)

# Reshape foi usado pois o modelo de LinearRegression exige que X seja uma matrix
# bidimensional e como X_bmi era unidimensional, ela teve que ser transformada
```

```python
import matplotlib.pyplot as plot
plot.scatter(X_bmi, y)
plot.ylabel("Blood Glucose (mg/dl)")
plot.xlabel("Body Mass Index")
plot.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_bmi, y)
predictions = reg.predict(X_bmi)

plot.scatter(X_bmi, y)
plot.plot(X_bmi, predictions, color="red")
plot.ylabel("Blood Glucose (mg/dl)")
plot.xlabel("Body Mass Index")
plot.show()
```
