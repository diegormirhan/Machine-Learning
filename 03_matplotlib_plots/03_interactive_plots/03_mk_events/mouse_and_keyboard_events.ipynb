```python
# Janela Qt: widgets precisam de backend interativo.
%matplotlib qt
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iris_dataset.csv").reset_index(drop=True)
fig, ax = plt.subplots(figsize=(9, 6))

sc = ax.scatter(
    df["petal length (cm)"],
    df["petal width (cm)"],
    c=pd.factorize(df["target"])[0],
    cmap="viridis",
    s=50,
    alpha=0.85,
    edgecolor="black",
    linewidths=0.4,
)

ax.set_title("Hover sobre os pontos para inspecionar", fontsize=14, pad=30)
ax.text(
    0.5,
    1.02,
    "Evento motion_notify_event + tooltip dinâmico",
    transform=ax.transAxes,
    ha="center",
    va="bottom",
    fontsize=10,
    color="#555555",
)
ax.set_xlabel("Petal length (cm)")
ax.set_ylabel("Petal width (cm)")

tooltip = ax.annotate(
    "",
    xy=(0, 0),
    xytext=(15, 15),
    textcoords="offset points",
    bbox=dict(boxstyle="round", fc="#fff8c5", ec="#aaaaaa"),
    arrowprops=dict(arrowstyle="->"),
)
tooltip.set_visible(False)


def on_hover(event):
    if event.inaxes != ax:
        return
    cont, ind = sc.contains(event)
    if cont:
        i = ind["ind"][0]
        x, y = sc.get_offsets()[i]
        tooltip.xy = (x, y)
        tooltip.set_text(f"#{i}\n{df.loc[i, 'target']}\n" f"petal: {x:.1f} × {y:.1f}")
        tooltip.set_visible(True)
    else:
        tooltip.set_visible(False)
    fig.canvas.draw_idle()


fig.canvas.mpl_connect("motion_notify_event", on_hover)
plt.savefig("mk_event.png", dpi=1000)
plt.show()
```
