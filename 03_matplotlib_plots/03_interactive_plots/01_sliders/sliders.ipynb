```python
# Backend interativo em janela separada (Qt); necessário para o slider responder ao arraste.
# Em alguns ambientes só há "inline"; aí troque por %matplotlib widget ou remova a linha.
%matplotlib qt
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider  # controle deslizante embutido no canvas
import numpy as np
```

```python
# Domínio em radianos: dois períodos da senoide quando ω=1 (sin(x) de 0 a 2π).
x = np.linspace(0, 2 * np.pi, 400)

fig, ax = plt.subplots(figsize=(9, 5))
# Margem inferior: área do slider. Margem superior: título + texto em coordenadas do eixo (y>1).
plt.subplots_adjust(bottom=0.22, top=0.84)
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

# Vírgula em (linha,): unpack de tupla com um único Line2D retornado por plot.
(linha,) = ax.plot(x, np.sin(x), color="#1f6dc7", linewidth=2)

ax.set_title(
    "Onda senoidal interativa — ajuste a frequência",
    fontsize=14,
    pad=30,
    fontweight="bold",
)
# transAxes: (0,0) canto inf.esq. do eixo, (1,1) sup.dir.; y>1 coloca o texto acima da área do gráfico.
ax.text(
    0.5,
    1.05,
    "Arraste o slider para ver o efeito em tempo real",
    transform=ax.transAxes,
    ha="center",
    va="bottom",
    fontsize=10,
    color="#555555",
)
ax.set_xlabel("x")
ax.set_ylabel("sin(ω·x)")
for s in ["top", "right"]:
    ax.spines[s].set_visible(False)

# Novo eixo invisível só para hospedar o widget: retângulo [esq, baixo, largura, altura] na figura (0–1).
ax_slider = plt.axes([0.18, 0.08, 0.65, 0.04])
slider_freq = Slider(
    ax=ax_slider,
    label="Frequência (ω)",
    valmin=0.5,
    valmax=10.0,
    valinit=1.0,
    valstep=0.1,
)


def atualizar(val):
    # val é ignorado aqui; o valor atual vem sempre de slider_freq.val após o movimento.
    linha.set_ydata(np.sin(slider_freq.val * x))
    # draw_idle: agenda redesenho no próximo ciclo do GUI (mais suave que draw() em loop).
    fig.canvas.draw_idle()


# Registra o callback: qualquer mudança no slider chama atualizar.
slider_freq.on_changed(atualizar)
plt.savefig(
    "sliders.png",
    dpi=1000,
    bbox_inches="tight",  # inclui títulos/anotações que ficam fora da caixa do eixo
    pad_inches=0.35,
)
plt.show()
```
