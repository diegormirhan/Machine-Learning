```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("iris_dataset.csv")
df.head()

```

```python
features = [
    "sepal length (cm)",
    "sepal width (cm)",
    "petal length (cm)",
    "petal width (cm)",
]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Distribuições das features - Iris', fontsize=15)

for ax, feat in zip(axes.ravel(), features):
    for species, group in df.groupby('target'):
        ax.hist(group[feat], bins=15, alpha=0.6, label=species)
    ax.set_title(feat)
    ax.set_xlabel(feat)
    ax.set_ylabel('Freq.')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig('subplots_iris.png', dpi=1000)
plt.show()
```
