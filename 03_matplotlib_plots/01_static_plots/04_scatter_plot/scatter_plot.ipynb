```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("iris_dataset.csv")
df.head()
```

```python
colors = {'setosa': '#1f77b4', 'versicolor': '#2ca02c', 'virginica': '#d62728'}

fig, ax = plt.subplots(figsize=(9, 6))

for species, group in df.groupby('target'):
    ax.scatter(group['petal length (cm)'], group['petal width (cm)'],
                    label=species, color=colors[species], alpha=0.7, edgecolor='black', linewidths=0.5
    )

ax.set_title('Iris - Comprimento x Largura da pétala', fontsize=14, pad=30)
ax.text(
    0.5,
    1.02,
    "Fonte: Iris Petal Types Historical Data",
    transform=ax.transAxes,
    ha="center",
    va="bottom",
    fontsize=10,
    color="#555555",
)

ax.set_xlabel('Comprimento da pétala (cm)')
ax.set_ylabel('Largura d apétala (cm)')
ax.set_xlim(0, 8)
ax.set_ylim(0, 3)

for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

for spine in ["left", "bottom"]:
    ax.spines[spine].set_color("#cccccc")

ax.grid(alpha=0.3)
ax.set_axisbelow(True)

ax.legend()
plt.tight_layout()
plt.savefig('scatter_plot_iris.png', dpi=1000)
plt.show()
```
