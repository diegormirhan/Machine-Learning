```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import pandas as pd
```

```python
df = pd.read_csv('house_prices.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(df['SalePrice'], bins=40, color='#9467bd', edgecolor='white')
ax.set_title('Distribuição do preço de venda de imóveis', fontsize=14, pad=30)
ax.text(0.5, 1.02, 'Fonte: House Prices Historical Data', transform=ax.transAxes, ha='center', va='bottom', fontsize=10, color='#555555')

ax.set_xlabel('Preço (USD)')
ax.set_ylabel('Frequência')
ax.set_ylim(0, 250)

ax.axvline(df['SalePrice'].mean(), color='red', linestyle='--', label=f"Média: ${df['SalePrice'].mean()/1000:,.0f}k")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: f"{x/1000:.0f}k"))

for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)

for spine in ["left", "bottom"]:
    ax.spines[spine].set_color("#cccccc")

ax.grid(
    axis="y",
    color="#e6e6e6",
    linewidth=0.8,
)
ax.set_axisbelow(True)

ax.legend()
plt.tight_layout()
plt.savefig('histogram_plot.png', dpi=1000)
plt.show()
```
