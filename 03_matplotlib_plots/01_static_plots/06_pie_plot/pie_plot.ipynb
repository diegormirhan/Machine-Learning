```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
counts = df['survived'].value_counts()

fig, ax = plt.subplots(figsize=(9, 6))

ax.pie(counts, autopct='%1.1f%%', colors=['#d62728', '#2ca02c'],
       explode=(0.02, 0.05), startangle=90)

ax.set_title('Proporção de sobreviventes - Titanic', fontsize=14, pad=30)
ax.text(
    0.5,
    1.02,
    "Fonte: Titanic Historical Data",
    transform=ax.transAxes,
    ha="center",
    va="bottom",
    fontsize=10,
    color="#555555",
)

plt.legend(['Não sobreviveu', 'Sobreviveu'])
plt.tight_layout()
plt.savefig('pie_plot_titanic.png', dpi=1000)
plt.show()
```
