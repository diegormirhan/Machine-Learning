```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
surv = df.groupby("pclass")["survived"].mean() * 100

surv.head()
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
bars = ax.bar(
    ["1ª Classe", "2ª Classe", "3ª Classe"],
    surv.values,
    color=["#2ca02c", "#ff7f0e", "#d62728"])
ax.set_title('Taxa de sobrevivência por classe no Titanic', fontsize=14, loc='center', pad=30)
ax.text(0.5, 1.02, 'Fonte: Titanic Historial Data', transform=ax.transAxes,  ha='center', va='bottom', fontsize=10, color='#555555')
ax.set_ylabel('Sobreviventes (%)')
ax.set_ylim(0, 100)

# anota valores no topo de cada barra
for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, h + 1, f'{h:.1f}%', ha='center')

for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

for spine in ['left', 'bottom']:
    ax.spines[spine].set_color('#cccccc')

ax.grid(axis='y', color='#e6e6e6', linewidth=0.8, )
ax.set_axisbelow(True)

plt.savefig("bar_chart_titanic.png", dpi=1000)
plt.tight_layout()
plt.show()
```
