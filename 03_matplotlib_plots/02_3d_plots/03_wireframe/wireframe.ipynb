```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401
import numpy as np
```

```python
# Grade no plano (x, y): 40 pontos por eixo (wireframe fica mais leve com menos amostras que uma superfície cheia).
x1d = np.linspace(-5, 5, 40)
y1d = np.linspace(-5, 5, 40)

# X e Y: matrizes com o mesmo formato; em (i,j) temos um par (x, y) do retângulo [-5,5] × [-5,5].
X, Y = np.meshgrid(x1d, y1d)

# Função z = f(x, y) = sin( √(x² + y²) )
# - √(x² + y²) é a distância r do ponto (x,y) até a origem, no plano horizontal (como o raio em polares).
# - sin(r) só depende dessa distância, não do ângulo → o gráfico tem simetria circular (ondas concêntricas).
# - É a versão 2D do padrão "seno de r", usado em livros para ilustrar ondas radiais / simetria esférica.
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='#1f6dc7', linewidth=0.6, rstride=2, cstride=2)

ax.set_title("Wireframe — sin(√(x² + y²))", fontsize=14, pad=20)
ax.text2D(
    0.5,
    1,
    "Onda radial — útil para entender propagação de sinais",
    transform=ax.transAxes,
    ha="center",
    va="bottom",
    fontsize=10,
    color="#555555",
)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.view_init(elev=30, azim=-55)

plt.tight_layout()
plt.savefig("wireframe_wave.png", dpi=1000)
plt.show()
```
