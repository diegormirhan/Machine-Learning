```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import numpy as np
```

```python
# Domínio quadrado no plano (x, y): 80 amostras por eixo.
x1d = np.linspace(-3, 3, 80)
y1d = np.linspace(-3, 3, 80)

# Malha: em cada (i, j), X[i,j] e Y[i,j] são as coordenadas onde avaliamos f(x, y).
X, Y = np.meshgrid(x1d, y1d)

# f(x, y) = e^{-(x² + y²)} · cos(2x)
# - e^{-(x²+y²)}: envelope gaussiano — pico em (0,0) e decaimento radial;
# - cos(2x): ondulação ao longo de x (período π em x).
# Parênteses em -(X**2 + Y**2) são obrigatórios: -(X**2) + (Y**2) seria e^{-x²+y²}, função diferente.
Z = np.exp(-(X**2 + Y**2)) * np.cos(2 * X)
```

```python
# contour3D desenha curvas no espaço onde Z é constante (isolinhas / "fatias" de altura fixa),
# em vez de preencher triângulos como plot_surface — ajuda a ver vales e cristas sem oclusão.
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection="3d")

# levels=40 → até ~40 valores de corte em Z; mais níveis = mais linhas, mais detalhe (e mais custo).
ax.contour3D(X, Y, Z, levels=40, cmap="plasma")

ax.set_title("Contour 3D — e^(−x²−y²) · cos(2x)", fontsize=14, pad=20)
ax.text2D(
    0.5,
    1,
    "Curvas de nível revelam estrutura sem oclusão",
    transform=ax.transAxes,
    ha="center",
    va="bottom",
    fontsize=10,
    color="#555555",
)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.view_init(elev=35, azim=-60)

plt.tight_layout()
plt.savefig("contour3d.png", dpi=1000)
plt.show()
```
