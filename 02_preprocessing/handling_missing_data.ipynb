```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("titanic.csv")
# print(df.isna().sum())
df = df.drop(columns=["cabin", "embarked", "fare", "boat", "body", "home.dest"])
# print(df.isna().sum())

X_age = df["age"].values.reshape(-1, 1)
print(X_age.shape)

X_not_age = df.drop(columns=["age", "survived"], axis=1).values
y = df["survived"].values

X_train_age, X_test_age, y_train_age, y_test_age = train_test_split(X_age, y, test_size=0.2, random_state=42)
X_train_not_age, X_test_not_age, y_train_not_age, y_test_not_age = train_test_split(X_not_age, y, test_size=0.2, random_state=42)

impute_age = SimpleImputer(strategy="most_frequent")
X_train_age = impute_age.fit_transform(X_train_age)
X_test_age = impute_age.transform(X_test_age)

print(X_train_age.shape, X_train_not_age.shape)
print(X_test_age.shape, X_test_not_age.shape)

X_train = np.append(X_train_age, X_train_not_age, axis=1)
X_test = np.append(X_test_age, X_test_not_age, axis=1)

print(X_train.shape, X_test.shape)
```
